All this infromation is based on the data provided by **Ofgem**

- **Number of Measures Installed** : It is a count of the number of energy efficiency measures installed in a household under GBIS.
- GBIS is designed to target a wide range of households in a **general eligibility group** as well as households in a **low-income group**.
- Households who receive specific benefits (or if in receipt of Child Benefit and are below the equivalised income threshold for the household composition) are members of the low-income group.
- The **Flexible Eligibility** component in the GBIS (Great Britain Insulation Scheme) gives Local Authorities (LAs) (like your city or council) the power to decide which households need help the most.
- The **Innovation Measure component** in the GBIS (Great Britain Insulation Scheme) is a way to encourage energy suppliers to try out new and better energy-saving solutions, like improved insulation or heating systems.These new measures are aimed specifically at : Low-income households, or Social housing with poor energy efficiency.
- It is possible for a measure delivered through Flexible Eligibility to also be an Innovation Measure.

### Measure types

- The below code cleans and organizes data before analysis.
- It identifies main categories by comparing the number of installations in each row to the total installations in related subcategories.
- Rows are labeled with the correct category to create clear and organized groupings.
- The main categories are determined by finding rows where the number of installations matches the total number of installations in smaller subcategories.
- Smaller subcategories are grouped under their respective main category.

```python
i = 0
while i < len(measure_types):
    current_value = measure_types.loc[i, "Measures Installed"]
    cumulative_sum = 0
    main_category = True

    # Cumulatively sum of rows
    for j in range(i + 1, len(measure_types)):
        cumulative_sum += measure_types.loc[j, "Measures Installed"]

        # Checking if the sum matches the current row's value
        if cumulative_sum == current_value:
            main_category = True
            measure_types.loc[i, "Main_Category"] = True
            measure_types.loc[i:j, "Category"] = measure_types.loc[i, "Measure Type"]  # Assign category
            break

        # If cumulative sum exceeds the main category's value then I am breaking the statment
        if cumulative_sum > current_value:
            main_category = False
            break

    # If no subcategories matched it will still treat as a main category
    if main_category:
        measure_types.loc[i, "Main_Category"] = True
        measure_types.loc[i, "Category"] = measure_types.loc[i, "Measure Type"]

    # Move to the next unmatched row
    i = j + 1 if main_category else i + 1

measure_types['Main_Category'] = measure_types['Main_Category'].fillna(False).astype(bool)
```

```python
# Filtering only the main categories

main_categories_df = measure_types[measure_types['Main_Category'] == True]
```

```python
# Listing the sub categories in each main category
```

```python
for index, row in main_categories_df.iterrows():
    main_category = row["Measure Type"]
    category_name = row["Category"]  

    
    subcategories_df = measure_types[(measure_types["Category"] == category_name) & (measure_types["Main_Category"] == False)]


    print(f"Main Category: {main_category}")
    for index, sub_row in subcategories_df.iterrows():
        print(f"  -- {sub_row['Measure Type']}")
    print() 
```
